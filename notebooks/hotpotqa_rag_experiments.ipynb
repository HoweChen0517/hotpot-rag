# HotpotQA RAG Experiments

This notebook demonstrates the LangChain-based HotpotQA RAG baseline and the core experiment flows.

```python
from hotpot_rag.data import load_hotpotqa_split
from hotpot_rag.experiments import ExperimentConfig, run_experiment
from hotpot_rag.indexing import build_corpus
from hotpot_rag.retrieval import build_bm25_retriever
```

```python
# loding a small sample from dataset and test retriever
samples = load_hotpotqa_split(config='distractor', split='validation', sample_size=5, seed=42)
documents = build_corpus(samples)
retriever = build_bm25_retriever(documents, top_k=5)
samples[0].question, retriever.retrieve(samples[0].question, top_k=3)[0].metadata
```
